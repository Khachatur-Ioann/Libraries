```python
import pandas as pd
```

```python
authors = pd.DataFrame({
    "author_id": [1, 2, 3],
    "author_name": ['Тургенев', 'Чехов', 'Островский']
})

authors
```

```python
book = pd.DataFrame({
    "author_id": [1, 1, 1, 2, 2, 3, 3],
    "book_title": ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
    "price": [450, 300, 350, 500, 450, 370, 290]
})

book
```

```python
df1 = pd.merge(authors, book, on='author_id', how='inner')

df1
```

```python
df1.nlargest(5, "price")
```

```python
authors_stat_1 = df1.groupby('author_name').agg({'price': 'max'}).rename(columns={'price':'max_price'})

authors_stat_1
```

```python
authors_stat_2 = df1.groupby('author_name').agg({'price': 'min'}).rename(columns={'price':'min_price'})

authors_stat_2
```

```python
authors_stat_3 = df1.groupby('author_name').agg({'price': 'mean'}).rename(columns={'price':'mean_price'})

authors_stat_3
```

```python
authors_stat = pd.concat([authors_stat_1, authors_stat_2, authors_stat_3], axis=1)

authors_stat
```
